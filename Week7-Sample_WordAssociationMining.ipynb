```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = '/Users/xniu2/Desktop/PyData/HealthNews' 
filelists = PlaintextCorpusReader(corpus_root, '.*',encoding='latin-1')
filelists.fileids()
```

```python
#create a list of news articles
news = []
for fileid in filelists.fileids():
    news.append(filelists.raw(fileid))
```

```python
news
```

```python
#import the shorttext library for text preprocessing
#make sure to have shorttext installed before importing. To install shorttext, run "pip install -U shorttext" at your command terminal
#in addition, in order to use shorttext in this case, you need to download a package "en" using the command "python -m spacy download en" at your command terminal
#standard_text_preprocessor_1 under shorttext.utils provides a standar way of text preprocessing, including the following steps:

 #removing special characters,
 #removing numerals,
 #converting all alphabets to lower cases,
 #removing stop words, and
 #stemming the words (using Porter stemmer).

import shorttext
preprocess = shorttext.utils.standard_text_preprocessor_1()
corpus = [preprocess(article).split(' ') for article in news]
```

```python
#convert the corpus of news to a document term matrix (dtm), where each row represents a document, each column represents the number of occurence for a word
dtm = shorttext.utils.DocumentTermMatrix(corpus, docids = filelists.fileids())
```

```python
# define a function to calculate entropy given a distribution
import numpy as np
import math

def entropy(p):
    if sum(p) == 0:
        return 0
    
    p = p/sum(p)

    p = p[ p > 0 ]
    
    H = -sum(p*np.log2(p))
    
    return H
```

```python
#print the number of "Dentistry" documents that contain the word "associ"
sub = 'Dentistry'
count_dentistry = 0
for item in associ_docs:
    if sub in item:
        count_dentistry += 1
print(count_dentistry)
```

```python
#print the number of "Depression" documents that contain the word "associ"
sub = 'Depression'
count_depression = 0
for item in associ_docs:
    if sub in item:
        count_depression += 1
print(count_depression)
```

```python
#print the number of "Diabetes" documents that contain the word "associ"
sub = 'Diabetes'
count_diabetes = 0
for item in associ_docs:
    if sub in item:
        count_diabetes += 1
print(count_diabetes)
```

```python
#this is an important step
#make an array, rows represent "Dentistry", "Depression", and "Diabetes" respectively. Columns represent the number of documents that contains the word "associ" and the number of documents that do NOT contain the word "associ"
array = np.reshape((count_dentistry,60-count_dentistry,count_depression,40-count_depression,count_diabetes,100-count_diabetes ),(3,2))
```

```python
array
```

```python
#calculate the column sum of the array
np.sum(array, axis = 0)
```

```python
#calculate the row sum of the array
np.sum(array, axis = 1)
```

```python
#calculate the entropy of the document class (Dentistry, Depression, and Diabetes)
marginal_entropy = entropy(np.sum(array, axis = 1))
```

```python
#calculate the column probabilities
column_probs = np.sum(array, axis = 0)/200
```

```python
column_probs
```

```python
#calculate the column entropies
column_entropy = np.apply_along_axis(entropy, 0, array)
```

```python
column_entropy
```

```python
conditional_entropy = sum(column_probs*column_entropy)
```

```python
conditional_entropy
```

```python
#calculate the mutual information between the word "associ" and the document class (Dentistry, Depression, and Diabetes)
MI_associ = marginal_entropy - conditional_entropy
```
