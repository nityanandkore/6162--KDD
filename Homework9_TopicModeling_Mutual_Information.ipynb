```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = '6162-KDD/DataSet/MovieReviews' 
filelists = PlaintextCorpusReader(corpus_root, '.*',encoding='latin-1')
filelists.fileids()
```

```python
#create a list of news articles
news = []
for fileid in filelists.fileids():
    news.append(filelists.raw(fileid))
```

```python
#convert the corpus of news to a document term matrix (dtm), where each row represents a document, each column represents the number of occurence for a word
dtm = shorttext.utils.DocumentTermMatrix(corpus, docids = filelists.fileids())
```

```python
# define a function to calculate entropy given a distribution
import numpy as np
import math

def entropy(p):
    if sum(p) == 0:
        return 0
    
    p = p/sum(p)

    p = p[ p > 0 ]
    
    H = -sum(p*np.log2(p))
    
    return H
```

```python
#print the number of Berardinelli reviews that contain the word 
import re
sub = 9999
count_Berardinelli= []
count_B = 0
for word_and_file in each_word_docs:
    #print(word_and_file[0])
    for file in word_and_file[1]:
      filelist= [int(i) for i in file.split('.') if i.isdigit()]
      #print(filelist)
      for x in filelist:
        if sub >= x:
          count_B += 1
    count_Berardinelli.append([word_and_file[0],count_B])
    count_B=0
    
count_Berardinelli
```

```python
#print the number of "Schwartz" documents that contain the word "director"

import re
sub = 9999
Count_Schwartz= []
count_S = 0
for word_and_file in each_word_docs:
    #print(word_and_file[0])
    for file in word_and_file[1]:
      filelist= [int(i) for i in file.split('.') if i.isdigit()]
      #print(filelist)
      for x in filelist:
        if sub <= x:
          count_S += 1
    Count_Schwartz.append([word_and_file[0],count_S])
    count_S=0

Count_Schwartz
```

```python
#this is an important step
#make an array, rows represent "Berardinelli" and "Schwartz" respectively. Columns represent the number of documents 
#that contains the word and the number of documents that do NOT contain the word 
array=[]
i=0
for word in count_Berardinelli:
  array.append(np.reshape((count_Berardinelli[i][1],80-count_Berardinelli[i][1],Count_Schwartz[i][1],100-Count_Schwartz[i][1],),(2,2)))
  i=i+1
```

```python
marginal_entropy = entropy(np.sum(array[0], axis = 1))

marginal_entropy
```

```python
#calculate the entropy of the document class (Berardinelli, Schwartz)
marginal_entropy=[]
i=0
for item in array:
  marginal_entropy.append(entropy(np.sum(array[i], axis = 1)))
  i+=1

marginal_entropy
```

```python
#calculate the column probabilities
column_probs=[]
i=0
for item in array:
  column_probs.append(np.sum(array[i], axis = 0)/180)
  i+=1

column_probs
```

```python
#calculate the column entropies
column_entropy=[]
i=0
for item in array:
  column_entropy.append(np.apply_along_axis(entropy, 0, array[i]))
  i+=1


column_entropy 
```

```python
conditional_entropy=[]
i=0
for item in array:
  conditional_entropy.append(sum(column_probs[i]*column_entropy[i]))
  i+=1

conditional_entropy 
```

```python
len(conditional_entropy)
```

```python
#calculate the mutual information between the word "director" and the document class (Berardinelli, Schwartz)
MI_associ=[]
i=0
for item in array:
  MI_associ.append(marginal_entropy[i] - conditional_entropy[i])
  i+=1

MI_associ 
```
