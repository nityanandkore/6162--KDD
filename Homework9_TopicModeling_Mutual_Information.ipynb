```python
from nltk.corpus import PlaintextCorpusReader

filelists_Berardinelli = PlaintextCorpusReader(corpus_Berardinelli, '.*',encoding='latin-1')
filelists_Berardinelli.fileids()
```

```python
#create a list of news articles
news_Berardinelli = []
for fileid in filelists_Berardinelli.fileids():
    news_Berardinelli.append(filelists_Berardinelli.raw(fileid))
```

```python
#import the shorttext library for text preprocessing
#make sure to have shorttext installed before importing. To install shorttext, run "pip install -U shorttext" at your command terminal
#in addition, in order to use shorttext in this case, you need to download a package "en" using the command "python -m spacy download en" at your command terminal
#standard_text_preprocessor_1 under shorttext.utils provides a standar way of text preprocessing, including the following steps:

 #removing special characters,
 #removing numerals,
 #converting all alphabets to lower cases,
 #removing stop words, and
 #stemming the words (using Porter stemmer).
#!pip install shorttext
import shorttext
preprocess = shorttext.utils.standard_text_preprocessor_1()
corpus = [preprocess(article).split(' ') for article in news_Berardinelli]
```

```python
#convert the corpus of news to a document term matrix (dtm), where each row represents a document, each column represents the number of occurence for a word
dtm = shorttext.utils.DocumentTermMatrix(corpus, docids = filelists_Berardinelli.fileids())
```

```python
# define a function to calculate entropy given a distribution
import numpy as np
import math

def entropy(p):
    if sum(p) == 0:
        return 0
    
    p = p/sum(p)

    p = p[ p > 0 ]
    
    H = -sum(p*np.log2(p))
    
    return H
```

```python
#print the number of "Berardinelli" documents that contain the word "director"
sub = 9999
count_Berardinelli = 0
for item in director_docs:
    filelist= [int(i) for i in item.split('.') if i.isdigit()]
    for x in filelist:
        if sub >= x:
         count_Berardinelli += 1
print(count_Berardinelli)
```

```python
#print the number of "Schwartz" documents that contain the word "director"
sub = 9999
count_Schwartz = 0
for item in director_docs:
    filelist= [int(i) for i in item.split('.') if i.isdigit()]
    for x in filelist:
        if sub < x:
         count_Schwartz += 1
print(count_Schwartz)
```

```python
#this is an important step
#make an array, rows represent "Berardinelli" and "Schwartz" respectively. Columns represent the number of documents 
#that contains the word "director" and the number of documents that do NOT contain the word "director"
array = np.reshape((count_Berardinelli,80-count_Berardinelli,count_Schwartz,100-count_Schwartz,),(2,2))
```

```python
array
```

```python
#calculate the column sum of the array
np.sum(array, axis = 0)
```

```python
#calculate the row sum of the array
np.sum(array, axis = 1)
```

```python
#calculate the entropy of the document class (Berardinelli, Schwartz)
marginal_entropy = entropy(np.sum(array, axis = 1))
```

```python
#calculate the column probabilities
column_probs = np.sum(array, axis = 0)/180
```

```python
column_probs
```

```python
#calculate the column entropies
column_entropy = np.apply_along_axis(entropy, 0, array)
```

```python
column_entropy
```

```python
conditional_entropy = sum(column_probs*column_entropy)
```

```python
conditional_entropy
```

```python
#calculate the mutual information between the word "director" and the document class (Berardinelli, Schwartz)
MI_associ = marginal_entropy - conditional_entropy
```
